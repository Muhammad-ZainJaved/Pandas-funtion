```python
import pandas as pd
```

```python
df = pd.read_csv('IMDB-Movie-Data.csv', index_col='Title')
```

```python
df.head()
```

```python
def rating_fun(x):
    if x >= 8.1:
        return 'Good'
    else:
        return 'Bad'
```

```python
df['Rating_c'] = df['Rating'].apply(rating_fun)
df.head()
```

```python
df['Rating_cc'] = df['Rating'].apply(lambda x: 'Good' if x >= 8.1 else 'Bad') # lambda expression
df.head()
```
